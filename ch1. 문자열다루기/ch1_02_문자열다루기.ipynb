## 추가, 분리, 정렬

```python
# 추가
address_words = ['서울특별시', '강서구', '방화동', '국립국어원']
address_words.append('1층')
address_words
```

```python
# 분리
address = "서울특별시 강서구 방화동 국립국어원"
words = address.split()
words
```

```python
# 순방향 정렬
address_words = ['서울특별시', '강서구', '방화동', '국립국어원']
address_words.sort()
address_words
```

```python
# 역방향 정렬
address_words = ['서울특별시', '강서구', '방화동', '국립국어원']
address_words.reverse()
address_words
```

```python
# 슬라이싱 [::-1]을 사용한 역방향 정렬
address_words[::-1]
```

```python
# join()은 분리했던 텍스트를 다시 원래대로 텍스트 형태로 복원
" ".join(address_words)
```

```python
# join()을 사용해 문자열 형태로 변경
"-".join(address_words)
```

## 문자열 내장 메서드 목록

* 언더스코어 두개(__)로 감싸져 있는 메서드는 매직 메서드 또는 스페셜 메서드라고 한다.
* 이 경우, 문자열.메서드명()이 아니라 스페셜 메시드명(문자열) 형태로 사용한다.

```python
print(dir(address))
```

```python
for func in dir(address):
    if not func.startswith("__"):
        print(func, end = ", ")
```
